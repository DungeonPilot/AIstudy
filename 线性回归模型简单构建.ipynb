```python
import tensorflow as tf
import random
import numpy as np
```

```python
# # 自拟输入数据
# print( random.randint(1,10) )        # 产生 1 到 10 的一个整数型随机数  
# print( random.random() )             # 产生 0 到 1 之间的随机浮点数
# print( random.uniform(1.1,5.4) )     # 产生  1.1 到 5.4 之间的随机浮点数，区间可以不是整数
# print( random.choice('tomorrow') )   # 从序列中随机选取一个元素
# print( random.randrange(1,100,2) )   # 生成从1到100的间隔为2的随机整数
rdm=np.random.RandomState(1)
X = rdm.rand(1000,2)
Y = np.array([x1*2+x2*3+random.random() for (x1,x2) in X])
Y = Y.reshape(1000,1)
print(Y.shape)
```

```python
#for name, age in zip(name_list, age_list): 同步遍历多个列表
```

```python
for x,y in zip(X,Y):
    print(x,y)
```

```python
x = tf.placeholder(tf.float32,shape=(None,2),name='x')
_y = tf.placeholder(tf.float32,shape=(None,1),name='y')

w = tf.Variable(tf.random_normal([2,1],stddev=1))
b = tf.Variable(tf.constant(0.0))
```

```python

y = tf.matmul(x,w)+b
```

```python
loss = tf.reduce_mean(tf.square(_y -  y))
```

```python
train_op = tf.train.AdadeltaOptimizer(0.1).minimize(loss)
```

```python
with tf.Session() as sess:
    losses=[]
    init = tf.initialize_all_variables()#写在 session 里
    sess.run(init)
    STEPS = 50000
    batchsize=50
    datasetsize = 1000
    for i in range(STEPS):
        start = (i*batchsize)%datasetsize
        end = min(start+batchsize,datasetsize)
        sess.run(train_op,feed_dict={x:X[start:end],_y:Y[start:end]})
        if i%1000 ==0:
            total_loss = sess.run(loss,feed_dict={x:X,_y:Y})
            losses.append(total_loss)
            print("at step %d loss = %f"%(i,total_loss))
    print(sess.run(w))
    print(sess.run(b))
```
